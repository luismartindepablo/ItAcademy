```python
import csv
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium.webdriver import Chrome
```

# Exercici 1

```python
# Download web
URL = 'https://www.bolsamadrid.es/esp/aspx/Portada/Portada.aspx'
```

```python
# Navigate to web
browser = Chrome("./chromedriver")
browser.get(URL)
```

```python
# Get the page
browser.find_element_by_link_text("Acciones").click()
soup = BeautifulSoup(browser.page_source, 'html.parser')
```

```python
# Select inf 
raw_table = soup.find("table", {"class": "TblPort", "id": "ctl00_Contenido_tblAcciones"})
rows = raw_table.find_all("tr")
```

```python
# Save in csv
table = open('table.csv', mode='w')
table_writer = csv.writer(table, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
```

```python
# Headers
headers = [x.text for x in rows[0].find_all("th")]
table_writer.writerow(headers);
```

```python
# Save row inf
for row in rows[1:]:
    columns = [x.text for x in row.find_all("td")]
    table_writer.writerow(columns)
```

```python
# Close CSV
table.close()
```
