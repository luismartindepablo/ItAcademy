### Exercici 1

```python
# Simple Random Sampling (100)
size = 100 # Tamaño de la muestra
random_sample = df_fifa.sample(size, replace=False) # Random sample
display(random_sample.head(10))
```

```python
# Systematic Sampling (uno de cada XX)
XX = 100 # Intervalo 
df_fifa_desordenado = df_fifa.sample(frac=1) # Mezclamos las filas
first_individual = random.randint(0,XX-1) # Primero aleatorio
systematic_sample = df_fifa_desordenado.iloc[first_individual::100] # Muestra
# Si queremos una fracción cortamos la muestra
# Al estar mezclada no influye en el resultado
size = 100 # Tamaño de la muestra
systematic_sample = systematic_sample.head(size) # Muestra
display(systematic_sample.head(10))
```

### Exercici 2

```python
# Stratified Sampling (by Nationality - 200)
size = 200 # Tamaño de la muestra
Nationalities = list(df_fifa.Nationality.unique()) # Categorias
stratified_sample = pd.DataFrame() # Muestra vacía
for Nat in Nationalities:
    df_temp = df_fifa[df_fifa.Nationality == Nat]
    percentage = len(df_temp)/len(df_fifa) # Presencia de la categoria en la población
    quantity = round(size * percentage) # Presencia de la categoria en la muestra
    stratified_sample = pd.concat([stratified_sample, df_temp.sample(quantity)])
display(stratified_sample.sample(10)) # Muestra de la muestra
```

```python
# Count por stratas
stratified_sample.groupby('Nationality').count().ID.head(10)
```

```python
# Tamaño
stratified_sample.shape
```

```python
# Stratified Sampling 2
size = 200; fraction = size/len(df_fifa)
stratified_sample = df_fifa.groupby('Nationality', group_keys = False).apply(pd.DataFrame.sample, frac=fraction)
stratified_sample.sample(10)
```

```python
# Count por stratas
stratified_sample.groupby('Nationality').count().ID.head(10)
```

```python
# Tamaño
stratified_sample.shape
```

```python
# SMOTE
# Mostramos la diferencia
minority = df_fifa[df_fifa.Preffered_Foot == "Left"]
majority = df_fifa[df_fifa.Preffered_Foot == "Right"]
print(len(minority), len(majority))
```

```python
# Seleccionamos aleatoriamente una muestra de la mayoria
majority_undersampling = majority.sample(len(minority))
# Creamos la muestra
undersampled_sample = pd.concat([minority, majority_undersampling])
undersampled_sample.sample(10)
```

```python
# Comprobamos la igualdad 
print(len(undersampled_sample[undersampled_sample.Preffered_Foot == "Left"]),
      len(undersampled_sample[undersampled_sample.Preffered_Foot == "Right"]))
```

### Exercici 3

```python
# Reservoir Sampling
size = 200; reservoir = [] # Tamaño de la muestra
for ii in range(len(df_fifa)):
    if len(reservoir) < size: # Llenamos la muestra de reserva
        reservoir.append(ii)
    else: # Cambiamos un elemento por otro con probabilidad p
        probability = size/(ii+1)
        if random.random() < probability:
             reservoir[random.choice(range(0,size))] = ii

reservoir_sample = df_fifa.iloc[reservoir] # Extraemos la muestra
reservoir_sample.head(10)
```
