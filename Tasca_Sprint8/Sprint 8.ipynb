### Exercici 1

**Hipotesis nula:** La media de edad de los jugadores es 25 años $\rightarrow H_0: \mu = 25 $

**Hipotesis alternativa:** La media de edad de los jugadores es diferente a 25 $\rightarrow H_1: \mu \neq 25 $

**Contraste de hipotesis sobre la media de una población con varianza desconocida:** T-student con n-1 grados de libertad. 

```python
# Age
Age = np.array(df_fifa.Age)
```

```python
# ----------------
# Test Estadístico
# ----------------

# Params
mu, alpha = 25, 0.05 

# Observed Value
mean = np.mean(Age) 
sd = np.std(Age)
n = len(Age)
tobs = (mean - mu) / (sd / np.sqrt(n))

# Acceptance Region
tcrit_L = t.ppf(alpha/2, n-1)
tcrit_U = t.ppf(1 - alpha/2, n-1)

# P value
pvalue = t.sf(abs(tobs), n-1) * 2

# Print
print(
    """
    tobs: {}
    tcrit_L: {}
    tcrit_U: {}
    pvalue: {}
    """
    .format(tobs, tcrit_L, tcrit_U, pvalue)
)
```

El valor observado cae fuera de la región de aceptación y el p-value es menor que el nivel de significancia. Se puede afirmar con un 95% de confianza que la media de edad de los jugadores es distinta a 25 años.

### Exercici 2

Se desea calcular si el porcentaje de jugadores con un Rating superior a 90 es diferente en Barcelona y Madrid.

**Hipotesis nula** $\rightarrow H_0: p_B = p_M$

**Hipotesis alternativa** $\rightarrow H_1: p_B \neq p_M $

**Contraste de hipotesis sobre la proporción de dos poblaciónes independientes:** Test estadístico N(0,1). 

```python
# Población Barcelona
df_fifa_B = df_fifa[df_fifa.Club == "FC Barcelona"].Rating
# Población Barcelona
df_fifa_M = df_fifa[df_fifa.Club == "Real Madrid"].Rating
```

```python
# ----------------
# Test Estadístico
# ----------------

# Params
alpha = 0.05

# Observed Value
nB = len(df_fifa_B); pB = sum( df_fifa_B>90 ) / nB
nM = len(df_fifa_M); pM = sum( df_fifa_M>90 ) / nM

p = (nB*pB + nM*pM) / (nB + nM)
zobs = (pB - pM) / np.sqrt( p*(1-p) * (1/nB + 1/nM) )

# Acceptance Region
zcritL = norm.ppf(alpha/2)
zcritU = norm.ppf(1 - alpha/2)

# P value
pvalue = norm.sf(abs(zobs))*2

# Print
print(
    """
    zobs: {}
    zcritL: {}
    zcritU: {}
    pvalue: {}
    """
    .format(zobs, zcritL, zcritU, pvalue)
)
```

El valor observado cae dentro de la región de aceptación y el p-value es mayor que el nivel de significancia. Con un 95% de confianza no se puede afirmar que el porcentaje de jugadores con un Rating superior a 90 es diferente en Barcelona y Madrid.

### Exercici 3

Queremos saber si los jugadores zurdos tienen mejor valoración que aquellos jugadores de campo diestros que tienen un peso, altura y edad similar. Como la muestra no está balanceada buscamos el jugador diestro más parecido a cada jugador zurdo.

**Hipotesis nula** $\rightarrow H_0: \mu_{DIESTROS} - \mu_{ZURDOS} = 0$

**Hipotesis alternativa** $\rightarrow H_1: \mu_{DIESTROS} - \mu_{ZURDOS} \neq 0$

**Contraste de hipotesis sobre la media de dos poblaciónes emparejadas:** T-student sobre la diferencia de medias muestrales con n-1 grados de libertad. 

```python
# -----------
# Muestras
# -----------

# Separamos los porteros
df_fifa_noGK = df_fifa[df_fifa.Club_Position != "GK"]

# Separamos diestros y zurdos
df_fifa_diestros = df_fifa_noGK[df_fifa_noGK.Preffered_Foot == "Right"]
df_fifa_zurdos = df_fifa_noGK[df_fifa_noGK.Preffered_Foot == "Left"]

# Muestras independientes
sample_diestros = df_fifa_diestros[["Age", "Weight", "Height"]]
sample_zurdos = df_fifa_zurdos[["Age", "Weight", "Height"]]
```

```python
# Distancia euclidea entre dos vectores
def euclidean( x1, x2 ):
    return ( np.sqrt( sum ( (x1-x2)**2 ) ) )

# Dado un jugador zurdo qual es el jugador diestro más similar 
def my_nn( x, sample ):
    dist = sample.apply(lambda y: euclidean(y, x), axis = 1)
    index = dist.idxmin()
    return(index)

# Dada la muestra de jugadores zurdos qual es la muestra de jugadores diestros correspondiente
def my_nn_sample( sample1, sample2 ):
    index_vec = sample1.apply(lambda y: my_nn(y, sample2), axis = 1)
    Right_paired = df_fifa.iloc[index_vec]
    return(Right_paired)
```

```python
# Encontramos la muestra emparejada (usamos 100 zurdos y 400 diestros pk es un código costoso)
Right_paired = my_nn_sample( sample_zurdos.iloc[0:100], sample_diestros.iloc[0:400] )

# Calculamos la diferencia
Rating_diff = np.array(df_fifa_zurdos.Rating.iloc[0:100]) - np.array(Right_paired.Rating)
```

```python
# ----------------
# Test Estadístico
# ----------------

# Params
alpha = 0.05

# Observed Value
mean = np.mean(Rating_diff) 
sd = np.std(Rating_diff)
n = len(Rating_diff)
tobs = mean / (sd / np.sqrt(n))

# Acceptance Region
#tcrit_L = t.ppf(alpha/2, n-1)
tcrit_U = t.ppf(1 - alpha/2, n-1)

# P value
pvalue = t.sf(abs(tobs), n-1) #* 2

# Print
print(
    """
    tobs: {}
    tcrit_L: {}
    tcrit_U: {}
    pvalue: {}
    """
    .format(tobs, tcrit_L, tcrit_U, pvalue)
)
```

El valor observado cae dentro de la región de aceptación y el p-value es mayor que el nivel de significancia. Con un 95% de confianza no se puede afirmar que la valoración media de los jugadores zurdos es distinta a la de los jugadpres diestros con su misma edad y constitución.
