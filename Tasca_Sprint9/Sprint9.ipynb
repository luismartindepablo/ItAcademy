```python
import pandas as pd
pd.options.display.max_columns = None
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, RobustScaler
```

```python
# Cargamos el dataset
df_fifa = pd.read_csv("fifa_clean.csv", encoding = "ISO-8859-1")
```

```python
# Muestra del dataset
display(df_fifa.head(10))
```

```python
# Estructura del Dataset
df_fifa.info()
```

```python
# Seleccionamos la columnas para el ejercicio
columnas = ["Club_Position", "Rating", "Height", "Weight", "Age", "Preffered_Foot"]
df_fifa = df_fifa[columnas]
```

```python
# Muestra del dataset
display(df_fifa.head(10))
```

```python
# Estructura del Dataset
df_fifa.info()
```

### Ejercicio 1

```python
# Dummy de Club_Position
Club_Position_Dummy = pd.get_dummies(df_fifa["Club_Position"])

# Dummy de Prefered_Food
Preffered_Food_Dummy = pd.get_dummies(df_fifa["Preffered_Foot"])

# Concat data
df_fifa = pd.concat([df_fifa, Club_Position_Dummy, Preffered_Food_Dummy], axis=1)
```

```python
# Inicializar Scaler
Scaler = StandardScaler()

#Escalar Numeric Variables
new_columns = ["Rating_scaled", "Height_scaled", "Weight_scaled", "Age_scaled"]
df_fifa[new_columns] = Scaler.fit_transform(df_fifa[["Rating", "Height", "Weight", "Age"]].values)
```

```python
# Print
df_fifa.head()
```

### Ejercicio 2

```python
# DataFrame for PCA
df_fifa_pca = pd.concat([df_fifa[new_columns], Club_Position_Dummy, Preffered_Food_Dummy], axis=1)
```

```python
# PCA de 2 componentes
pca = PCA(n_components=2)
modelo_pca = pca.fit_transform(df_fifa_pca)

# Registros con las nuevas variables
display(pd.DataFrame(data = modelo_pca, columns = ['PC1', 'PC2']).head())

# Componentes en función de las viejas variables
display(pd.DataFrame(data=pca.components_, columns=df_fifa_pca.columns, index=["PC1", "PC2"]))

# Explained variance
var = pca.explained_variance_ratio_.round(2)
print('''
La primera componente explica un {} de la varianza,
y la segunda un {}.

En total, el modelo conserva el {} de varianza
'''.format(var[0], var[1], sum(var)))
```

### Ejercicio 3

```python
# Escalar Robusto
Scaler = RobustScaler()

#Escalar Numeric Variables
new_columns = ["Rating_robust", "Height_robust", "Weight_robust", "Age_robust"]
df_fifa[new_columns] = Scaler.fit_transform(df_fifa[["Rating", "Height", "Weight", "Age"]].values)
```
